## Bitwise Operations and Masking

To demonstrate these operations let's create some simple images

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2 
from google.colab import drive
drive.mount('/content/drive')
from google.colab.patches import cv2_imshow
```

```python

# Making a sqare
square=np.zeros((300,300),dtype=np.uint8)
cv2.rectangle(square,(50, 50), (250, 250), 255, -2)
cv2_imshow(square)
cv2.waitKey(0)

# Making a ellipse
ellipse=np.zeros((300,300),dtype=np.uint8)
cv2.ellipse(ellipse, (150, 150), (150, 150), 30, 0, 180, 255, -1)
cv2_imshow( ellipse)
cv2.waitKey(0)

cv2.destroyAllWindows()
```

### Experimenting with some bitwise operations

```python
# Shows only where they intersection
AND=cv2.bitwise_and(square,ellipse)
cv2_imshow(AND)
cv2.waitKey(0)

# Shows only where they union
OR=cv2.bitwise_or(square,ellipse)
cv2_imshow(OR)
cv2.waitKey(0)

# Not show intersection
XOR=cv2.bitwise_xor(square,ellipse)
cv2_imshow(XOR)
cv2.waitKey(0)

# Shows everything that isn't part of the square
NOT=cv2.bitwise_not(square,ellipse)
cv2_imshow(NOT)
cv2.waitKey(0)
```
