```python
import numpy as np
import matplotlib.pyplot as plt
import cv2 
from google.colab import drive
drive.mount('/content/drive')
from google.colab.patches import cv2_imshow
```

```python
image=cv2.imread("/content/drive/Shareddrives/mayur/computer vision/images/gradient.jpg")
cv2_imshow(image)
cv2.waitKey(0)
```

```python
# Values below 127 goes to 0 (black, everything above goes to 255 (white)
ret,thresh1=cv2.threshold(image,127,255,cv2.THRESH_BINARY)
cv2_imshow(thresh1)
cv2.waitKey(0)
```

```python
# Values below 127 go to 255 and values above 127 go to 0 (reverse of above)
ret,thresh2=cv2.threshold(image,127,255,cv2.THRESH_BINARY_INV)
cv2_imshow(thresh2)
cv2.waitKey(0)
```

```python

# Values above 127 are truncated (held) at 127 (the 255 argument is unused)
ret,thresh3=cv2.threshold(image,127,255,cv2.THRESH_TRUNC)
cv2_imshow(thresh3)
cv2.waitKey(0)
```

```python
# Values below 127 go to 0, above 127 are unchanged  
ret,thresh4 = cv2.threshold(image, 127, 255, cv2.THRESH_TOZERO)
cv2_imshow(thresh4)
```

```python
# Resever of above, below 127 is unchanged, above 127 goes to 0
ret,thresh5 = cv2.threshold(image, 127, 255, cv2.THRESH_TOZERO_INV)
cv2_imshow(thresh5)
cv2.waitKey(0) 
    
cv2.destroyAllWindows()
```

### Is there a better way off thresholding?

The biggest downfall of those simple threshold methods is that we need to provide the threshold value (i.e. the 127 value we used previously).
#### What if there was a smarter way of doing this?

There is with, Adaptive thresholding. 

```python
img=cv2.imread('/content/drive/Shareddrives/mayur/computer vision/images/Origin_of_Species.jpg')
cv2_imshow(img)
```

```python
# Values below 127 goes to 0 (black, everything above goes to 255 (white)
ret,thresh1=cv2.threshold(img,127,255,cv2.THRESH_BINARY)
cv2_imshow(thresh1)
```

```python
# It's good practice to blur images as it removes noise
image=cv2.blur(img,(3,3),0)
cv2_imshow(image)
```

```python



# Using adaptiveThreshold
thresh = cv2.adaptiveThreshold(img, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, 3, 5) 
cv2_imshow(thresh) 
cv2.waitKey(0) 

_, th2 = cv2.threshold(image, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)
cv2_imshow(thresh) 
cv2.waitKey(0) 
```
