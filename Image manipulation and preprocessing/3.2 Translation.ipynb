## Translations

This an affine transform that simply shifts the position of an image.

We use cv2.warpAffine to implement these transformations.

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2 
from google.colab import drive
drive.mount('/content/drive')
from google.colab.patches import cv2_imshow
```

```python
image=cv2.imread("/content/drive/Shareddrives/mayur/computer vision/images/input.jpg")

# store height and width
heigth,width=image.shape[:2]

quarter_heigth,quarter_width=heigth/4,width/4

#       | 1 0 Tx |
#  T  = | 0 1 Ty |

# T is our translation matrix
T=np.float32([[1,0,quarter_width],[0,1,quarter_heigth]])

# We use warpAffine to transform the image using the matrix, T
img_translation=cv2.warpAffine(image,T,(width,heigth))
cv2_imshow(img_translation)
cv2.waitKey()
cv2.destroyAllWindows()
```
