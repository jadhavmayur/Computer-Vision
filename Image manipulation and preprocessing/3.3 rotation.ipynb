## Rotations

cv2.getRotationMatrix2D(rotation_center_x, rotation_center_y, angle of rotation, scale)


```python
import numpy as np
import matplotlib.pyplot as plt
import cv2 
from google.colab import drive
drive.mount('/content/drive')
from google.colab.patches import cv2_imshow
```

```python
img=cv2.imread("/content/drive/Shareddrives/mayur/computer vision/images/input.jpg")

heigth,width=img.shape[:2]

rotation_matrix=cv2.getRotationMatrix2D((width/2,heigth/2),90,0.5)

rotated_image=cv2.warpAffine(img,rotation_matrix,(width,heigth))
cv2_imshow(rotated_image)
cv2.waitKey()
cv2.destroyAllWindows()
```

#transpose

```python
transpose=cv2.transpose(img)
cv2_imshow(transpose)
cv2.waitKey()
cv2.destroyAllWindows()
```

#flip

```python
flipped=cv2.flip(transpose,1)
cv2_imshow(flipped)
cv2.waitKey()
cv2.destroyAllWindows()
```
